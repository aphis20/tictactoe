```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import pandas as pd

# Placeholder functions for WEPP and Tillage Erosion models
def wepp_model(soil_type, slope, rainfall, crop_cover):
    # Replace with actual model logic
    erosion_rate = slope * rainfall * 0.1 * (1 - crop_cover/100)  # Simplified formula
    return erosion_rate

def tillage_erosion_model(tillage_intensity, slope):
    # Replace with actual model logic
    tillage_erosion = tillage_intensity * slope * 0.05  # Simplified formula
    return tillage_erosion

# Function to run the models and update the GUI
def run_models():
    try:
        # Validate inputs
        slope = float(slope_var.get())
        rainfall = float(rainfall_var.get())
        crop_cover = float(crop_cover_var.get())
        tillage_intensity = float(tillage_intensity_var.get())
        
        if not (0 <= slope <= 100):
            raise ValueError("Slope must be between 0 and 100.")
        if not (0 <= crop_cover <= 100):
            raise ValueError("Crop cover must be between 0 and 100.")
        
        soil_type = soil_type_var.get()
        wepp_result = wepp_model(soil_type, slope, rainfall, crop_cover)
        tillage_result = tillage_erosion_model(tillage_intensity, slope)
        
        # Display results
        wepp_label.config(text=f"WEPP Erosion Rate: {wepp_result:.2f} tons/ha/year")
        tillage_label.config(text=f"Tillage Erosion: {tillage_result:.2f} tons/ha/year")
        
        # Plot the results
        plot_results(wepp_result, tillage_result)
        
    except ValueError as e:
        messagebox.showerror("Input Error", str(e))

# Function to plot the results using matplotlib
def plot_results(wepp_result, tillage_result):
    fig, ax = plt.subplots(figsize=(8, 6))
    
    categories = ['WEPP Erosion', 'Tillage Erosion']
    values = [wepp_result, tillage_result]
    
    ax.bar(categories, values, color=['blue', 'orange'])
    ax.set_ylabel('Erosion (tons/ha/year)')
    ax.set_title('Erosion Model Results')
    
    # Clear previous plot and display the new one
    for widget in plot_frame.winfo_children():
        widget.destroy()
    
    canvas = FigureCanvasTkAgg(fig, master=plot_frame)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Function to export the results and graphs to files
def export_results():
    file_path = filedialog.asksaveasfilename(defaultextension=".png",
                                             filetypes=[("PNG files", "*.png"),
                                                        ("All files", "*.*")])
    if file_path:
        fig = plt.figure(figsize=(8, 6))
        
        categories = ['WEPP Erosion', 'Tillage Erosion']
        values = [float(wepp_label.cget("text").split(": ")[1].split()[0]), 
                  float(tillage_label.cget("text").split(": ")[1].split()[0])]
        
        plt.bar(categories, values, color=['blue', 'orange'])
        plt.ylabel('Erosion (tons/ha/year)')
        plt.title('Erosion Model Results')
        plt.savefig(file_path)
        messagebox.showinfo("Export Successful", f"Results exported to {file_path}")

# Function to customize model parameters (Advanced Options)
def customize_model():
    custom_window = tk.Toplevel(app)
    custom_window.title("Customize Model Parameters")
    
    tk.Label(custom_window, text="Advanced Model Parameters").grid(row=0, column=0, columnspan=2)
    
    tk.Label(custom_window, text="Custom Parameter 1:").grid(row=1, column=0)
    custom_param1 = tk.StringVar(value="Default Value 1")
    tk.Entry(custom_window, textvariable=custom_param1).grid(row=1, column=1)
    
    tk.Label(custom_window, text="Custom Parameter 2:").grid(row=2, column=0)
    custom_param2 = tk.StringVar(value="Default Value 2")
    tk.Entry(custom_window, textvariable=custom_param2).grid(row=2, column=1)
    
    def apply_custom_params():
        # Apply the custom parameters to the models
        print(f"Applied {custom_param1.get()} and {custom_param2.get()} to the models")
        custom_window.destroy()
    
    tk.Button(custom_window, text="Apply", command=apply_custom_params).grid(row=3, column=0, columnspan=2)

# Main Tkinter application window
app = tk.Tk()
app.title("Advanced Erosion Model App")

# Input fields
tk.Label(app, text="Soil Type:").grid(row=0, column=0)
soil_type_var = tk.StringVar(value="Loam")
ttk.Combobox(app, textvariable=soil_type_var, values=["Loam", "Clay", "Sand"]).grid(row=0, column=1)

tk.Label(app, text="Slope (%):").grid(row=1, column=0)
slope_var = tk.StringVar(value="5")
tk.Entry(app, textvariable=slope_var).grid(row=1, column=1)

tk.Label(app, text="Rainfall (mm/year):").grid(row=2, column=0)
rainfall_var = tk.StringVar(value="800")
tk.Entry(app, textvariable=rainfall_var).grid(row=2, column=1)

tk.Label(app, text="Crop Cover (%):").grid(row=3, column=0)
crop_cover_var = tk.StringVar(value="75")
tk.Entry(app, textvariable=crop_cover_var).grid(row=3, column=1)

tk.Label(app, text="Tillage Intensity:").grid(row=4, column=0)
tillage_intensity_var = tk.StringVar(value="1")
tk.Entry(app, textvariable=tillage_intensity_var).grid(row=4, column=1)

# Buttons
tk.Button(app, text="Run Models", command=run_models).grid(row=5, column=0, pady=10)
tk.Button(app, text="Export Results", command=export_results).grid(row=5, column=1, pady=10)
tk.Button(app, text="Customize Model", command=customize_model).grid(row=6, column=0, columnspan=2, pady=10)

# Labels to display the results
wepp_label = tk.Label(app, text="WEPP Erosion Rate: ")
wepp_label.grid(row=7, column=0, columnspan=2)

tillage_label = tk.Label(app, text="Tillage Erosion: ")
tillage_label.grid(row=8, column=0, columnspan=2)

# Frame to hold the plot
plot_frame = tk.Frame(app)
plot_frame.grid(row=9, column=0, columnspan=2)

# Start the Tkinter main loop
app.mainloop()
```
