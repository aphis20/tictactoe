```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value1': np.random.randn(100),
    'Value2': np.random.randn(100),
    'Value3': np.random.randn(100)
})
```

```python
# Line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Value1'], label='Value1')
plt.plot(data['Value2'], label='Value2')
plt.title('Line Plot of Value1 and Value2')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value1', data=data, ci=None)
plt.title('Bar Plot of Value1 by Category')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Category', data=data)
plt.title('Scatter Plot of Value1 vs Value2 by Category')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
rainfall_intensity = np.linspace(10, 100, 10)  # in mm/hr
soil_erosion_rate = np.random.uniform(5, 25, 10)  # in tons/ha

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(rainfall_intensity, soil_erosion_rate, marker='o')
plt.title('Effect of Rainfall Intensity on Soil Erosion Rate')
plt.xlabel('Rainfall Intensity (mm/hr)')
plt.ylabel('Soil Erosion Rate (tons/ha)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Crop Cover': ['No Cover', 'Partial Cover', 'Full Cover'] * 10,
    'Soil Loss': np.random.uniform(10, 50, 30)
})

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Crop Cover', y='Soil Loss', data=data)
plt.title('Impact of Crop Cover on Soil Retention')
plt.xlabel('Crop Cover')
plt.ylabel('Soil Loss (tons/ha)')
plt.show()
```

```python
# Sample data
slope_gradient = np.linspace(5, 45, 10)  # in degrees
sediment_yield = np.random.uniform(15, 60, 10)  # in tons/ha

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(slope_gradient, sediment_yield, c='blue')
plt.title('Influence of Slope Gradient on Sediment Yield')
plt.xlabel('Slope Gradient (degrees)')
plt.ylabel('Sediment Yield (tons/ha)')
plt.grid(True)
plt.show()
```
